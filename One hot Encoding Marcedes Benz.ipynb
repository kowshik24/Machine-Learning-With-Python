```python
import pandas as pd
import numpy as np

data = pd.read_csv(r'F:\Machine Learning\Kaggle\Marcedes Benz\train.csv')
```

```python
# let's find how many unique category in each and every column

for col in data.columns:
    print(col, " " ,len(data[col].unique()) , 'labels ')
```

```python
# let's examine how many columns we will obtain after one hot encodin these varibales
pd.get_dummies(data,drop_first=True).shape
```

```python
data.X2.value_counts().sort_values(ascending=False).head(20)
```

```python
# let's make a list with the most frequent categories of the variable

top_10 = [x for x in data.X2.value_counts().sort_values(ascending=False).head(10).index]
top_10
```

```python
# and now we make the 10 binary variables

for label in top_10:
    data[label] = np.where(data['X2']==label , 1 , 0)

data[['X2']+top_10].head(40)
```

```python
def one_hot_top_x(df,variable,top_x_labels):
    
    for  label in top_x_labels:
        df[variable+"_"+label] = np.where(df[variable]==label,1,0)
        

columns = []
for x in range(386):
    if 'X'+str(x) in data:
        if len(data['X'+str(x)].unique()) >= int(10):
            columns.append('X'+str(x))



data = pd.read_csv(r'F:\Machine Learning\Kaggle\Marcedes Benz\train.csv' , usecols =columns)

one_hot_top_x(data,'X2',top_10)

data.head()
```

```python
# More Generally we can create

columns = []
for x in range(386):
    if 'X'+str(x) in data:
        if len(data['X'+str(x)].unique()) >= int(10):
            columns.append('X'+str(x))

data = pd.read_csv(r'F:\Machine Learning\Kaggle\Marcedes Benz\train.csv' , usecols = ['X1','X2','X3','X4','X5','X6'])

def findTop10ValuesInEachColumns(data):
    for col in data.columns:
        top_10 = [x for x in data[col].value_counts().sort_values(ascending=False).head(10).index]
        one_hot_top_x(data,col,top_10)
        print(top_10)
        
findTop10ValuesInEachColumns(data)
```
