```python
import pandas as pd
import datetime
```

```python
df_base = datetime.datetime.today()
df_date_list = [df_base-datetime.timedelta(days=x) for x in range (0,20)]
df = pd.DataFrame(df_date_list)
df.columns = ['day']
df
```

```python
 df['day_of_week'] = df['day'].dt.weekday_name
df.head()
```

```python
weekday_map = {
    'Monday':1,
    'Tuesday':2,
    'Wednesday':3,
    'Thursday':4,
    'Friday':5,
    'Saturday':6,
    'Sunday':7
}
df['day_of_week'] = df.day_of_week.map(weekday_map)
df.head(10)
```
