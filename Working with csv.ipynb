# Sep Parameter

```python
pd.read_csv('all_data.csv',names=['Order ID','Product','Quantity Ordered','Price Each','Order Date','Purchase Address'],sep='')
```

```python
pd.read_csv('all_data.csv',index_col='Order ID')
```

# Header Parameter

```python
pd.read_csv('all_data.csv',header=0)
```

# use_cols parameters

```python
pd.read_csv('all_data.csv',usecols=['Product','Quantity Ordered','Price Each','Order Date','Purchase Address'])
```

# Squeeze parameter

```python
pd.read_csv('all_data.csv',usecols=['Product'],squeeze=True)
```

# Skiprows/nrows Parameter

```python
pd.read_csv('all_data.csv',skiprows=[0,1])
```

```python
pd.read_csv('all_data.csv',nrows=100)
```

# Encoding Parameters

```python
pd.read_csv('all_data.csv',encoding='UTF-8')
```

# Skip Bad Lines

```python
pd.read_csv('all_data.csv',sep=',',encoding='UTF-8',error_bad_lines=False)
```

# dtypes

```python
pd.read_csv('all_data.csv',dtype={'Quantity Ordered':int})
```

 # Handing Dates

```python
pd.read_csv('all_data.csv',parse_dates=['Order Date'])
```

```python
pd.read_csv('all_data.csv',parse_dates=['Order Date']).info()
```

# Converters

```python
def rename(name):
    if name == 'iPhone':
        return 'IPhone'
    else:
        return name
```

```python
pd.read_csv('all_data.csv',converters={'Product':rename})
```

# na_values

```python
pd.read_csv('all_data.csv',na_values=['iPhone'])
```
