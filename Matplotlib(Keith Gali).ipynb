```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# Basic Graph

```python
plt.plot([1,2,3],[2,4,6])
```

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

plt.plot(x,y,label='2x',color='red',linewidth=3,marker='.' ,linestyle='--', markersize = 15 ,markeredgecolor='blue')
x2 = np.arange(0,4.5,0.5)
plt.plot(x2,x2**2,'b',label = 'x^2')
plt.title('Our first Graph',fontdict={'fontname':'Comic Sans MS' , 'fontsize':20})
plt.xlabel('X axis',fontdict={'fontname':'Comic Sans MS'})
plt.ylabel('Y axis')
plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,7,8,10])
plt.legend()
plt.show()
```

```python
# use short hand notation
# fmt = '[color][marker][line]'

x = [0,1,2,3,4]
y = [0,2,4,6,8]

plt.plot(x,y,'b.--',label='2x')
plt.title('Our first Graph',fontdict={'fontname':'Comic Sans MS' , 'fontsize':20})
plt.xlabel('X axis',fontdict={'fontname':'Comic Sans MS'})
plt.ylabel('Y axis')
plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,7,8,10])
plt.legend()
plt.show()
```

```python
## line no 2
x2 = np.arange(0,4.5,0.5)
plt.plot(x,y,'b',label = '2x')
plt.plot(x2,x2**2,label='x^2',color='red',linewidth=3,marker='.',linestyle='--',markersize=5,markeredgecolor='blue')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
```

```python
#plt.plot(x,y,'r',label='2x')
x2 = np.arange(0,4.5,0.5)
plt.plot(x2[:6],x2[:6]**2,'b',label = '2x')
plt.plot(x2[5:],x2[5:]**2,'r--')
plt.show()
```

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

# resize the figure
plt.figure(figsize=(5,3), dpi=300)

plt.plot(x,y,label='2x',color='red',linewidth=3,marker='.' ,linestyle='--', markersize = 15 ,markeredgecolor='blue')
x2 = np.arange(0,4.5,0.5)
plt.plot(x2,x2**2,'b',label = 'x^2')
plt.title('Our first Graph',fontdict={'fontname':'Comic Sans MS' , 'fontsize':20})
plt.xlabel('X axis',fontdict={'fontname':'Comic Sans MS'})
plt.ylabel('Y axis')
plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,7,8,10])
plt.legend()
plt.show()
```

# Barchart

```python
labels = ['A' , 'B' , 'C']
values = [1,4,2]
plt.bar(values,labels)
plt.show()
```

```python
labels = ['A' , 'B' , 'C']
values = [1,4,2]

bars = plt.bar(labels,values)


bars[0].set_hatch('/')
bars[1].set_hatch('o')
bars[2].set_hatch('-')

plt.figure(figsize=(6,4))


plt.show()
```

```python

labels = ['A' , 'B' , 'C']
values = [1,4,2]

bars = plt.bar(labels,values)


patterns = ['/','o','*']

for bar in bars:
    bar.set_hatch(patterns.pop(0))

#bars[0].set_hatch('/')
#bars[1].set_hatch('o')
#bars[2].set_hatch('-')

plt.figure(figsize=(6,4))


plt.show()

```
