# Line Graph

```python
gas = pd.read_csv('gas_prices.csv')
```

```python
gas
```

```python
plt.figure(figsize=(8,5))
plt.title('Gas Prices Over Time in USD',fontdict={'fontweight':'bold','fontsize':18})
plt.plot(gas.Year,gas.USA,'b.-',label='United States')
plt.plot(gas['Year'],gas.Canada,'r.-',label='Canada')
plt.plot(gas.Year,gas['South Korea'],'g.-',label='South Korea')
plt.plot(gas.Year,gas.Australia,'y.-',label='Australia')
plt.xticks(gas.Year[::3])

plt.xlabel('Year')
plt.ylabel('US Dollars')
#print(gas.Year[0::3])

plt.legend()
plt.savefig('gas_price_figure.png',dpi=300)
plt.show()
```

```python
# Another way to plot!


plt.figure(figsize=(8,5))
plt.title('Gas Prices Over Time in USD')
plt.xlabel('Year')
plt.ylabel('US Dollar')
for country in gas:
    if country != 'Year':
        plt.plot(gas.Year,gas[country],marker='.',label=country)
        plt.xticks(gas.Year[::3])

plt.legend()
plt.show()
    
```

```python
fifa.Weight = [int(x.strip('lbs')) if type(x) == str else x for x in fifa.Weight]
```

```python
fifa.Weight[0]
```

```python
light = fifa.loc[fifa.Weight < 125].count()[0]
light_medium = fifa.loc[(fifa.Weight >= 125) & (fifa.Weight<150)].count()[0]
medium = fifa.loc[(fifa.Weight >= 150) & (fifa.Weight<175)].count()[0]
medium_heavy = fifa.loc[(fifa.Weight >= 175) & (fifa.Weight<200)].count()[0]
heavy = fifa.loc[fifa.Weight >= 200].count()[0]
print(light)
print(light_medium)
print(medium)
print(medium_heavy)
print(heavy)
```

```python
weights = [light,light_medium,medium,medium_heavy,heavy]

plt.style.use('ggplot')

labels = ['Under 125','125-150','150-175','175-200','Over 200']

explode = (.4,.2,0,0,.4)

plt.title('Weight Distribution of fifa in (lbs)')

plt.pie(weights,autopct='%.2f%%',labels=labels,pctdistance = 0.8,explode=explode)

plt.show()
```

```python
plt.style.use('default')

plt.figure(figsize=(5,8))
barcelona = fifa.loc[fifa.Club=='FC Barcelona']['Overall']
madrid = fifa.loc[fifa.Club=='Real Madrid']['Overall']
revs = fifa.loc[fifa.Club=='New England Revolution']['Overall']
labels = ['FC Barcelona' , 'Real Madrid' , 'NE Revoltion']
plt.title('Professional Soccer Team Comperision')
plt.ylabel('FIFA Overall Rating')
colors = ['red' , 'green' , 'blue']

boxes = plt.boxplot([barcelona,madrid,revs],labels = labels , patch_artist = True)
for box in boxes['boxes']:
    # Set edge color
    box.set(color='#2bcdef',linewidth = 2)
    
    # Change Fill Color
    box.set(facecolor='#3bcdef')
plt.show()
```
