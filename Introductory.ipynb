# Spliting and Merging Color Channels

```python
img_1 = cv2.imread("F:\AAA Machine Learning\OpenCV\img1.jpg",cv2.IMREAD_COLOR)

b , g , r = cv2.split(img_1)

plt.figure(figsize=[20,5])
plt.subplot(141);plt.imshow(r,cmap='gray');plt.title("Red Channel");
plt.subplot(142);plt.imshow(g,cmap='gray');plt.title("Green Channel");
plt.subplot(143);plt.imshow(r,cmap='gray');plt.title("Blue Channel");

imgMarged = cv2.merge((b,g,r))

plt.subplot(144);plt.imshow(imgMarged[:,:,::-1]);plt.title("Marged Output")

```

# Converting to Different Color Spaces

### Changing from BGR to RGB

```python
img_new_rgb = cv2.cvtColor(img_1,cv2.COLOR_BGR2RGB)
plt.imshow(img_new_rgb)
```

#### Changing to HSV color space

```python
img_hsv = cv2.cvtColor(img_1,cv2.COLOR_BGR2HSV)

h , s , v = cv2.split(img_hsv)

plt.figure(figsize=[20,5])
plt.subplot(141);plt.imshow(h,cmap='gray');plt.title("H Channel");
plt.subplot(142);plt.imshow(s,cmap='gray');plt.title("S Channel");
plt.subplot(143);plt.imshow(v,cmap='gray');plt.title("V Channel");
plt.subplot(144);plt.imshow(img_hsv);plt.title("Originial");
```

#### Modifying indivisual Channel

```python
h_new = h + 10
img_1_marged = cv2.merge((h_new,s,v))
img_1_rgb = cv2.cvtColor(img_1_marged,cv2.COLOR_HSV2RGB)


plt.figure(figsize=[20,5])
plt.subplot(141);plt.imshow(h,cmap='gray');plt.title("H Channel");
plt.subplot(142);plt.imshow(s,cmap='gray');plt.title("S Channel");
plt.subplot(143);plt.imshow(v,cmap='gray');plt.title("V Channel");
plt.subplot(144);plt.imshow(img_1_rgb);plt.title("Originial");
```

### Saving Images

```python
cv2.imwrite("F:\AAA Machine Learning\OpenCV\img2.jpg",img_1_rgb)

Image(filename="F:\AAA Machine Learning\OpenCV\img2.jpg")
```
